Soal No 2 A

```python
# a. Histogram for Purchase Price, Current Price, and Gain/Loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data=data, x='Purchase Price', kde=True, edgecolor='white', color='brown')
plt.title('Distribution of Purchase Price', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Menunjukan atribut pada Current Price
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 2)
sns.histplot(data=data, x='Current Price', kde=True, color='green', edgecolor='white')
plt.title('Distribution of Current Price')
plt.tight_layout()
plt.show()
```

```python
## Menunjukan atribut pada Gain/Loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data=data, x='Gain/Loss', kde=True, edgecolor='white', color='grey')
plt.tight_layout()
plt.show()
```

Soal No 2B

```python
# Membersihkan tanda mata uang atau spasi dalam kolom 'Dividend/share'
data['Dividend/share'] = data['Dividend/share'].apply(lambda x: str(x).replace('$', '').replace(',', '').strip())

# Konversi tipe data 'Dividend/share' ke float
data['Dividend/share'] = data['Dividend/share'].astype(float)

# Mengurutkan berdasarkan Dividend/share tertinggi
sorted_by_dividend = data.nlargest(5, 'Dividend/share')

# Menampilkan atribut Stock Name dan Symbol dengan Dividend/share tertinggi
print("Atribut Stock Name dan Symbol dengan Dividend/share 5 tertinggi:")
print(sorted_by_dividend[['Stock Name', 'Symbol']])
```

Soal No 2C

```python
# Membersihkan format khusus pada kolom 'Market Value'
data['Market Value'] = data['Market Value'].astype(str)
data['Market Value'] = data['Market Value'].str.replace('$', '').str.replace(',', '')

# Konversi tipe data 'Market Value' ke float
data['Market Value'] = data['Market Value'].astype(float)

# Mengurutkan berdasarkan Market Value tertinggi
sorted_by_market_value = data.nlargest(5, 'Market Value')

# Menampilkan atribut Stock Name dan Symbol dengan Market Value tertinggi
print("Atribut Stock Name dan Symbol dengan Market Value  5 tertinggi:")
print(sorted_by_market_value[['Stock Name', 'Symbol']])
```

Soal No 2D

```python
# Membersihkan format khusus pada kolom 'Annual Yield'
data['Annual Yield'] = data['Annual Yield'].astype(str).str.replace('%', '')

# Konversi tipe data 'Annual Yield' ke float
data['Annual Yield'] = data['Annual Yield'].astype(float)

# Mengurutkan berdasarkan Annual Yield tertinggi
sorted_by_annual_yield = data.nlargest(7, 'Annual Yield')

# Menampilkan atribut Stock Name dan Symbol dengan Annual Yield tertinggi
print("Atribut Stock Name dan Symbol dengan Annual Yield tertinggi:")
print(sorted_by_annual_yield[['Stock Name', 'Symbol']])
```
